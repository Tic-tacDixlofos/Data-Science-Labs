```python
import pandas as pd
import matplotlib.pyplot as plt

# File paths
file1 = "tur1.xlsx"
file2 = "tur2.xlsx"
file3 = "pop.xlsx"
file4 = "pop2.xlsx"
output_file = "united.xlsx"


df1 = pd.read_excel(file1)
df2 = pd.read_excel(file2)
df3 = pd.read_excel(file3)
df4 = pd.read_excel(file4)


united_df = pd.concat([df1, df2, df3, df4], ignore_index=True)


columns_of_interest = ['Stoianoglo \nAlexandr', 'Sandu \nMaia', 'Usatîi\nRenato', 'i)numărul buletinelor de vot primite\nde biroul electoral al secției\nde votare', 'Unnamed: 2']


if all(col in united_df.columns for col in columns_of_interest):

    united_df[columns_of_interest] = united_df[columns_of_interest].apply(pd.to_numeric, errors='coerce')


    united_df_cleaned = united_df.dropna(subset=columns_of_interest)


    correlation_matrix = united_df_cleaned[columns_of_interest].corr()


    print("Correlation Matrix:")
    print(correlation_matrix)


    corr_values = correlation_matrix.iloc[0, 1:]


    fig, ax = plt.subplots(figsize=(8, 6))
    wedges, texts, autotexts = ax.pie(corr_values, labels=corr_values.index, autopct='%1.1f%%', startangle=90,
                                      wedgeprops=dict(width=0.4), colors=['#66b3ff', '#99ff99', '#ffcc99'])


    centre_circle = plt.Circle((0, 0), 0.3, fc='white')
    ax.add_artist(centre_circle)


    ax.axis('equal')


    plt.title("Correlation Values - Donut Chart")


    plt.show()


    united_df.to_excel(output_file, index=False)
else:
    print("One or more specified columns are missing from the data.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.DataFrame({'Unnamed: 30': [1, 2], 'Unnamed: 31': [1, 2]})
data1 = pd.DataFrame({'Unnamed: 30': [2, 3], 'Unnamed: 31': [2, 3]})
data2 = pd.DataFrame({'Unnamed: 30': [3, 4], 'Unnamed: 31': [3, 4]})
data3 = pd.DataFrame({'Unnamed: 30': [4, 5], 'Unnamed: 31': [4, 5]})


united_df = pd.concat([data, data1, data2, data3], ignore_index=True)


print(united_df.columns)


united_df.rename(columns={'Unnamed: 30': 'Male', 'Unnamed: 31': 'Female'}, inplace=True)


columns_of_interest = ['Male', 'Female']
filtered_df = united_df[columns_of_interest]


summary = filtered_df.sum()


plt.figure(figsize=(6, 4))
summary.plot(kind='bar', color=['blue', 'pink'], alpha=0.8)
plt.title("Summary of Male and Female")
plt.xlabel("Gender")
plt.ylabel("Sum of Values")
plt.xticks(rotation=0)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


data = pd.DataFrame({
    'Localitatea': ['City A', 'City B', 'City C', 'City D'],
    'Stoianoglo \nAlexandr': [10, 20, 15, 5],
    'Sandu \nMaia': [30, 25, 35, 10],
    'Usatîi\nRenato': [5, 15, 10, 20]
})


plt.figure(figsize=(10, 6))


x_positions = np.arange(len(data['Localitatea']))


plt.scatter(
    x_positions,
    data['Stoianoglo \nAlexandr'],
    s=data['Stoianoglo \nAlexandr'] * 10,  # Bubble size
    color='blue',
    alpha=0.6,
    label='Stoianoglo Alexandr'
)
plt.scatter(
    x_positions,
    data['Sandu \nMaia'],
    s=data['Sandu \nMaia'] * 10,
    color='green',
    alpha=0.6,
    label='Sandu Maia'
)
plt.scatter(
    x_positions,
    data['Usatîi\nRenato'],
    s=data['Usatîi\nRenato'] * 10,
    color='red',
    alpha=0.6,
    label='Usatîi Renato'
)


plt.xticks(x_positions, data['Localitatea'], rotation=45, ha='right')
plt.title('Candidate Popularity by Locality')
plt.xlabel('Locality')
plt.ylabel('Votes')
plt.legend(title='Candidates', loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.DataFrame({
    'Localitatea': ['City A', 'City B', 'City C', 'City D'],
    'Stoianoglo \nAlexandr': [10, 20, 15, 5],
    'Sandu \nMaia': [30, 25, 35, 10],
    'Usatîi\nRenato': [5, 15, 10, 20]
})

data1 = pd.DataFrame({
    'Localitatea': ['City A', 'City B', 'City C', 'City D'],
    'Stoianoglo \nAlexandr': [12, 18, 17, 4],
    'Sandu \nMaia': [28, 30, 32, 12],
    'Usatîi\nRenato': [6, 16, 11, 18]
})


merged_df = pd.merge(data, data1, on='Localitatea', suffixes=('_data', '_data1'))


merged_df['Stoianoglo \nAlexandr Diff'] = merged_df['Stoianoglo \nAlexandr_data'] - merged_df['Stoianoglo \nAlexandr_data1']
merged_df['Sandu \nMaia Diff'] = merged_df['Sandu \nMaia_data'] - merged_df['Sandu \nMaia_data1']
merged_df['Usatîi\nRenato Diff'] = merged_df['Usatîi\nRenato_data'] - merged_df['Usatîi\nRenato_data1']


print(merged_df)


fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot for Stoianoglo Alexandr
merged_df[['Localitatea', 'Stoianoglo \nAlexandr_data', 'Stoianoglo \nAlexandr_data1']].set_index('Localitatea').plot(kind='bar', ax=axes[0], color=['blue', 'green'])
axes[0].set_title('Stoianoglo Alexandr Comparison')
axes[0].set_ylabel('Votes')
axes[0].set_xlabel('Locality')

# Plot for Sandu Maia
merged_df[['Localitatea', 'Sandu \nMaia_data', 'Sandu \nMaia_data1']].set_index('Localitatea').plot(kind='bar', ax=axes[1], color=['blue', 'green'])
axes[1].set_title('Sandu Maia Comparison')
axes[1].set_ylabel('Votes')
axes[1].set_xlabel('Locality')

# Plot for Usatîi Renato
merged_df[['Localitatea', 'Usatîi\nRenato_data', 'Usatîi\nRenato_data1']].set_index('Localitatea').plot(kind='bar', ax=axes[2], color=['blue', 'green'])
axes[2].set_title('Usatîi Renato Comparison')
axes[2].set_ylabel('Votes')
axes[2].set_xlabel('Locality')

plt.tight_layout()
plt.show()

# Plot Differences for each candidate
merged_df[['Localitatea', 'Stoianoglo \nAlexandr Diff', 'Sandu \nMaia Diff', 'Usatîi\nRenato Diff']].set_index('Localitatea').plot(kind='bar', figsize=(8, 6), color=['blue', 'green', 'red'])
plt.title('Vote Differences Between data and data1')
plt.ylabel('Vote Difference')
plt.xlabel('Locality')
plt.tight_layout()
plt.show()
```
